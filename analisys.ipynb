```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import plotly.graph_objects as go

df = pd.read_csv("out/llm_result.csv")

pg_colors = ['#003DA5', '#00A3E0', '#FFB81C', '#E4002B', '#78BE20']
sns.set_theme(style="whitegrid")
```

На какой позиции оказывается бренд.

```python
df['rank_label'] = df['a_pg_rank'].apply(
    lambda x: '1-е место' if x == 1 else ('В списке (не первый)' if x > 1 else 'Не упомянут'))

df_pct = df.groupby(['бренд', 'категория_вопроса', 'rank_label']).size().reset_index(name='count')
df_total = df.groupby(['бренд', 'категория_вопроса']).size().reset_index(name='total')
df_plot = df_pct.merge(df_total, on=['бренд', 'категория_вопроса'])
df_plot['percentage'] = (df_plot['count'] / df_plot['total']) * 100

fig = px.bar(df_plot,
             x="бренд",
             y="percentage",
             color="rank_label",
             facet_col="категория_вопроса",
             facet_col_wrap=2,
             title="<b>Эффективность присутствия P&G (в % от общего числа запросов)</b>",
             labels={'percentage': 'Доля ответов (%)', 'бренд': 'Бренд'},
             color_discrete_map={
                 '1-е место': '#003DA5',
                 'В списке (не первый)': '#00A3E0',
                 'Не упомянут': '#E0E0E0'
             },
             category_orders={
                 "rank_label": ["1-е место", "В списке (не первый)", "Не упомянут"],
                 "категория_вопроса": ["брендовые", "категориальные", "консультационные", "сравнительные"]
             },
             height=800)

fig.update_layout(
    yaxis_ticksuffix="%",
    legend_title="Позиция P&G",
    font=dict(size=12)
)

fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))

fig.show()
```

Как часто мы выигрываем рекомендацию по сравнению с внешними конкурентами.

```python
win_stats = df.groupby(['бренд', 'категория_вопроса'])[['a_is_winner', 'a_comp_is_winner']].mean().reset_index()

win_stats['a_is_winner'] *= 100
win_stats['a_comp_is_winner'] *= 100

win_melted = win_stats.melt(
    id_vars=['бренд', 'категория_вопроса'],
    value_vars=['a_is_winner', 'a_comp_is_winner'],
    var_name='WinnerType',
    value_name='WinRatePct'
)

win_melted['WinnerType'] = win_melted['WinnerType'].map({
    'a_is_winner': 'Победа P&G',
    'a_comp_is_winner': 'Победа Конкурента'
})

fig = px.bar(
    win_melted,
    x='бренд',
    y='WinRatePct',
    color='WinnerType',
    facet_col='категория_вопроса',
    facet_col_wrap=2,
    barmode='group',
    title="<b>Win Rate: Кто чаще становится 'Выбором Нейросети'?</b>",
    labels={'WinRatePct': 'Доля рекомендаций (%)', 'бренд': 'Бренд'},
    color_discrete_map={
        'Победа P&G': '#003DA5',
        'Победа Конкурента': '#E4002B'
    },
    category_orders={
        "категория_вопроса": ["брендовые", "категориальные", "консультационные", "сравнительные"]
    },
    height=800
)

fig.update_layout(
    yaxis_ticksuffix="%",
    legend_title="Тип рекомендации",
    margin=dict(t=100, b=50, l=50, r=50)
)

fig.for_each_annotation(lambda a: a.update(text=f"<b>{a.text.split('=')[-1].upper()}</b>"))

fig.show()
```

Глубина описания технологий и коммерческих атрибутов в ответах нейросети по каждому бренду.

```python
brand_stats = df.groupby('бренд').agg({
    'a_has_tech': 'mean',
    'a_has_price': 'mean',
    'a_buy_cta': 'mean',
    'запрос': 'count'
}).reset_index()

for col in ['a_has_tech', 'a_has_price', 'a_buy_cta']:
    brand_stats[col] = brand_stats[col] * 100
brand_stats['size_for_plot'] = brand_stats['a_buy_cta'] + 5
fig = px.scatter(
    brand_stats,
    x='a_has_tech',
    y='a_has_price',
    size='size_for_plot',
    color='бренд',
    hover_name='бренд',
    text='бренд',
    custom_data=['a_buy_cta'],
    title='<b>Глобальный профиль контента: Технологии vs Коммерция</b>',
    labels={
        'a_has_tech': 'Глубина описания технологий (%)',
        'a_has_price': 'Наличие цен и мест продаж (%)'
    },
    range_x=[25, 80],
    range_y=[0, 35],
    height=800,
    template='plotly_white',
    size_max=25)

fig.update_traces(
    hovertemplate='<b>%{hovertext}</b><br>Технологии: %{x:.1f}%<br>Цены: %{y:.1f}%<br>Реальный CTA: %{customdata[0]:.1f}%<extra></extra>',
    textposition='top center',
    marker=dict(line=dict(width=1.5, color='DarkSlateGrey'))
)

v_line = brand_stats['a_has_tech'].mean()
h_line = brand_stats['a_has_price'].mean()
fig.add_vline(x=v_line, line_dash='dot', line_color='rgba(0,0,0,0.2)')
fig.add_hline(y=h_line, line_dash='dot', line_color='rgba(0,0,0,0.2)')

fig.update_layout(
    title_x=0.5,
    xaxis_ticksuffix='%',
    yaxis_ticksuffix='%',
    showlegend=False)

fig.show()
```

Баланс тональности ответов нейросети по каждому бренду и типу вопроса.

```python
df['sentiment_label'] = df['a_sentiment'].map({
    1: 'Позитивный (Рекомендация)',
    0: 'Нейтральный (Факты)',
    -1: 'Негативный (Критика)'
})

sent_group = df.groupby(['бренд', 'категория_вопроса', 'sentiment_label']).size().reset_index(name='count')
sent_totals = df.groupby(['бренд', 'категория_вопроса']).size().reset_index(name='total')
sent_plot = sent_group.merge(sent_totals, on=['бренд', 'категория_вопроса'])
sent_plot['percentage'] = (sent_plot['count'] / sent_plot['total']) * 100

fig = px.bar(
    sent_plot,
    y="бренд",
    x="percentage",
    color="sentiment_label",
    facet_col="категория_вопроса",
    facet_col_wrap=2,
    orientation='h',
    title="<b>Sentiment Balance по категориям запросов</b>",
    labels={'percentage': 'Доля (%)', 'бренд': ''},
    color_discrete_map={
        'Позитивный (Рекомендация)': '#78BE20',
        'Нейтральный (Факты)': '#E0E0E0',
        'Негативный (Критика)': '#E4002B'
    },
    category_orders={
        "sentiment_label": ["Позитивный (Рекомендация)", "Нейтральный (Факты)", "Негативный (Критика)"],
        "категория_вопроса": ["брендовые", "категориальные", "консультационные", "сравнительные"]
    },
    height=900,
    template="plotly_white"
)

fig.update_layout(
    xaxis_ticksuffix="%",
    legend_title="Тональность",
    title_x=0.5,
    margin=dict(t=100, b=50)
)

fig.for_each_annotation(lambda a: a.update(text=f"<b>{a.text.split('=')[-1].upper()}</b>"))

fig.show()
```

Где нейросеть находит информацию о брендах P&G: распределение по типам источников.

```python
source_labels = {
    1: '1. Маркетплейсы',
    2: '2. Офиц. сайты',
    3: '3. Обзоры/Статьи',
    4: '4. Форумы/Q&A',
    5: '5. Аптеки',
    6: '6. Смешанные'
}
df['source_name'] = df['a_source_trust'].map(source_labels)

source_dist = df.groupby(['бренд', 'source_name']).size().reset_index(name='count')
source_totals = df.groupby('бренд').size().reset_index(name='total')
source_plot = source_dist.merge(source_totals, on='бренд')
source_plot['percentage'] = (source_plot['count'] / source_plot['total']) * 100

pivot_sources = source_plot.pivot(index='бренд', columns='source_name', values='percentage').fillna(0)

fig = px.imshow(
    pivot_sources,
    labels=dict(x="Тип источника", y="Бренд", color="Доля (%)"),
    x=pivot_sources.columns,
    y=pivot_sources.index,
    color_continuous_scale='Blues',
    title="<b>Где нейросеть находит информацию о ваших брендах? (%)</b>",
    text_auto=".1f",
    aspect="auto",
    height=600
)

fig.update_layout(
    title_x=0.5,
    xaxis_title="",
    yaxis_title="Бренд",
    coloraxis_colorbar=dict(title="% ответов")
)

fig.show()
```

Сводный дашборд по всем ключевым метрикам эффективности брендов в LLM-выдаче.

```python
summary = df.groupby('бренд').agg({
    'a_pg_rank': lambda x: (x == 1).mean() * 100,
    'a_is_winner': lambda x: x.mean() * 100,
    'a_sentiment': 'mean',
    'a_has_tech': lambda x: x.mean() * 100,
    'a_has_price': lambda x: x.mean() * 100,
    'a_buy_cta': lambda x: x.mean() * 100
}).reset_index().round(1)

summary = summary.sort_values(by='a_is_winner', ascending=False)


def get_sent_color(val):
    if val > 0.2: return 'rgba(120, 190, 32, 0.5)'
    if val < -0.1: return 'rgba(228, 0, 43, 0.5)'
    return 'rgba(224, 224, 224, 0.5)'


sent_colors = [get_sent_color(x) for x in summary['a_sentiment']]

fig = go.Figure(data=[go.Table(
    header=dict(
        values=['<b>БРЕНД</b>', '<b>1-Е МЕСТО (%)</b>', '<b>ПОБЕДА (%)</b>',
                '<b>SENTIMENT</b>', '<b>ТЕХНОЛОГИИ (%)</b>', '<b>ЦЕНЫ (%)</b>', '<b>CTA (%)</b>'],
        fill_color='#003DA5',
        align='center',
        font=dict(color='white', size=14),
        height=40
    ),
    cells=dict(
        values=[
            summary['бренд'],
            summary['a_pg_rank'].map('{:.1f}%'.format),
            summary['a_is_winner'].map('{:.1f}%'.format),
            summary['a_sentiment'],
            summary['a_has_tech'].map('{:.1f}%'.format),
            summary['a_has_price'].map('{:.1f}%'.format),
            summary['a_buy_cta'].map('{:.1f}%'.format)
        ],
        fill_color=[
            'white', 'white', 'white', sent_colors, 'white', 'white', 'white'
        ],
        align='center',
        font=dict(size=13),
        height=30
    )
)])

fig.update_layout(
    title="<b>Сводный дашборд эффективности брендов P&G в LLM-выдаче</b>",
    margin=dict(l=10, r=10, t=50, b=10),
    height=500
)

fig.show()
```
